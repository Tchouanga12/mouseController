```python
import os
import sys
import matplotlib.pyplot as plt
```

```python
precision_list=['INT8', 'FP16', 'FP32']

face_inference_time=[]
face_fps=[]
face_model_load_time=[]

landmark_inference_time=[]
landmark_fps=[]
landmark_model_load_time=[]

headpose_inference_time=[]
headpose_fps=[]
headpose_model_load_time=[]

gaze_inference_time=[]
gaze_fps=[]
gaze_model_load_time=[]

for precision in precision_list:
    with open('/Users/hephzibah/Documents/INTEL/starter/'+precision+'/face_stats.txt', 'r') as f:
        face_inference_time.append(float(f.readline().split("\n")[0]))
        face_fps.append(float(f.readline().split("\n")[0]))
        face_model_load_time.append(float(f.readline().split("\n")[0]))
    
    with open('/Users/hephzibah/Documents/INTEL/starter/'+precision+'/landmark_stats.txt', 'r') as f:
        landmark_inference_time.append(float(f.readline().split("\n")[0]))
        landmark_fps.append(float(f.readline().split("\n")[0]))
        landmark_model_load_time.append(float(f.readline().split("\n")[0]))
        
    with open('/Users/hephzibah/Documents/INTEL/starter/'+precision+'/headpose_stats.txt', 'r') as f:
        headpose_inference_time.append(float(f.readline().split("\n")[0]))
        headpose_fps.append(float(f.readline().split("\n")[0]))
        headpose_model_load_time.append(float(f.readline().split("\n")[0]))
        
    with open('/Users/hephzibah/Documents/INTEL/starter/'+precision+'/gaze_stats.txt', 'r') as f:
        gaze_inference_time.append(float(f.readline().split("\n")[0]))
        gaze_fps.append(float(f.readline().split("\n")[0]))
        gaze_model_load_time.append(float(f.readline().split("\n")[0]))
```

```python
plt.plot(precision_list, face_inference_time, 'r', marker='o', label='Face Detection Model')
plt.plot(precision_list, landmark_inference_time, 'b', marker='o', label='Landmark Detection Model')
plt.plot(precision_list, headpose_inference_time, 'g', marker='o', label='Head-pose Estimation Model')
plt.plot(precision_list, gaze_inference_time, 'm', marker='o', label='Gaze Detection Model')
plt.xlabel("Model Precision")
plt.ylabel("Inference Time in Seconds")
plt.title('Inference time of the models for different precisions',fontweight="bold")
plt.legend(loc='right')
plt.show()
```

```python
plt.plot(precision_list, face_fps, 'r', marker='o', label='Face Detection Model')
plt.plot(precision_list, landmark_fps, 'b', marker='o', label='Landmark Detection Model')
plt.plot(precision_list, headpose_fps, 'g', marker='o', label='Head-pose Estimation Model')
plt.plot(precision_list, gaze_fps, 'm', marker='o', label='Gaze Detection Model')
plt.xlabel("Model Precision")
plt.ylabel("FPS")
plt.title('Frames Per Second',fontweight="bold")
plt.legend(loc='right')
plt.show()
```

```python
plt.plot(precision_list, face_model_load_time, 'r', marker='o', label='Face Detection Model')
plt.plot(precision_list, landmark_model_load_time, 'b', marker='o', label='Landmark Detection Model')
plt.plot(precision_list, headpose_model_load_time, 'g', marker='o', label='Head-pose Estimation Model')
plt.plot(precision_list, gaze_model_load_time, 'm', marker='o', label='Gaze Detection Model')
plt.xlabel("Model Precision")
plt.ylabel("Model Load Time (s)")
plt.title('Time Taken to Load the Models',fontweight="bold")
plt.legend(loc='right')
plt.show()
```
